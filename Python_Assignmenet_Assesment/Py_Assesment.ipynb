```python
#Healthcare

class cliniAappoinment:
    def __init__(self):
        self.appoinments = []
        self.time_slots = ["10am","11am","12pm","2pm","3pm"]


    def book(self):
         name = input("Patient Name: ")
         age = int(input("Age: "))
         mobile=input("mobile number: ")
         doctor =input("Preferred Doctor: ")

         print("Time slots:" , self.time_slots)
         slot =input("Choose Slot: ")


         count=0
         for app in self.appoinments:
            if app["doctor"]==doctor and app["slot"]==slot:
                count+=1

         if count >=3:
          print("This slot is full for this doctor. ")
          return

         self.appoinments.append({
          "name":name,
          "age":age,
          "mobile":mobile,
          "doctor":doctor,
          "slot":slot
         })
         print("Appointment Booked! ")

    def view(self):
        mobile = input("Enter mobile number: ")
        fount =False
        for app in self.appoinments:
            if app["mobile"]==mobile:
                print(app)
                fount=True
        if not fount:
            print("No Appointment found")

    def cancel(self):
        mobile = input("Enter mobile number: ")
        for app in self.appoinments:
            if app["mobile"]==mobile:
                self.appoinments.remove(app)
                print("appoinment cancelled. ")
                return
            print("no appointment found")


clinic = cliniAappoinment()

while True:
    print("\n1. Book Appointment")
    print("2. View Appointment")
    print("3. Cancel Appointment")
    print("4. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        clinic.book()
    elif choice == "2":
        clinic.view()
    elif choice == "3":
        clinic.cancel()
    elif choice == "4":
        break
    else:
        print("Invalid choice.")
```

```python
#School Management

class SchoolManagement:
    def __init__(self):
        self.students = {}
        self.student_id = 1

    # Naya Admission
    def add_student(self):
        name = input("Student Name: ")
        age = int(input("Age: "))
        student_class = int(input("Class (1-12): "))
        mobile = input("Guardian Mobile (10 digits): ")

        # Age check
        if age < 5 or age > 18:
            print(" Age 5 se 18 ke beech honi chahiye.")
            return

        # Mobile check
        if len(mobile) != 10 or not mobile.isdigit():
            print(" Mobile 10 digit ka hona chahiye.")
            return

        # Record save karna
        self.students[self.student_id] = {
            "name": name,
            "age": age,
            "class": student_class,
            "mobile": mobile
        }
        print(f" Admission done! Student ID: {self.student_id}")
        self.student_id += 1

    # Student Details dekhna
    def view_student(self):
        sid = int(input("Enter Student ID: "))
        if sid in self.students:
            s = self.students[sid]
            print(f"ID: {sid}, Name: {s['name']}, Age: {s['age']}, Class: {s['class']}, Mobile: {s['mobile']}")
        else:
            print(" Student ID not found.")

    # Student update karna
    def update_student(self):
        sid = int(input("Enter Student ID: "))
        if sid in self.students:
            new_class = input("New Class (blank = skip): ")
            new_mobile = input("New Mobile (blank = skip): ")

            if new_class:
                self.students[sid]["class"] = int(new_class)

            if new_mobile:
                if len(new_mobile) == 10 and new_mobile.isdigit():
                    self.students[sid]["mobile"] = new_mobile
                else:
                    print(" Invalid mobile number.")
                    return

            print(" Record updated.")
        else:
            print(" Student ID not found.")

    # Student record delete karna
    def remove_student(self):
        sid = int(input("Enter Student ID: "))
        if sid in self.students:
            del self.students[sid]
            print(" Record deleted.")
        else:
            print(" Student ID not found.")


# ======= MAIN PROGRAM =======
system = SchoolManagement()

while True:
    print("\n--- School Management ---")
    print("1. New Admission")
    print("2. View Student")
    print("3. Update Student")
    print("4. Remove Student")
    print("5. Exit")
    ch = input("Enter choice: ")

    if ch == "1":
        system.add_student()
    elif ch == "2":
        system.view_student()
    elif ch == "3":
        system.update_student()
    elif ch == "4":
        system.remove_student()
    elif ch == "5":
        print(" Exit ho gaya.")
        break
    else:
        print(" Wrong choice, try again.")
```

```python
#BusReservation

class BusReservation:
    def __init__(self):

        self.routes = {
            1: {"route": "Mumbai to Pune", "price": 500, "seats": 40},
            2: {"route": "Delhi to Jaipur", "price": 600, "seats": 40},
            3: {"route": "Bangalore to Chennai", "price": 700, "seats": 40},
            4: {"route": "Ahmedabad to Surat", "price": 300, "seats": 40}
        }
        self.tickets = {}     # store ticket details
        self.ticket_id = 1    # unique ticket counter

    # Show available routes
    def show_routes(self):
        print("\n Available Routes:")
        for rid, info in self.routes.items():
            print(f"{rid}. {info['route']} - ‚Çπ{info['price']} (Seats left: {info['seats']})")

    # Book a ticket
    def book_ticket(self):
        self.show_routes()
        try:
            route_choice = int(input("Choose route number: "))
            if route_choice not in self.routes:
                print(" Invalid route.")
                return

            route_info = self.routes[route_choice]
            if route_info["seats"] <= 0:
                print(" No seats left on this route.")
                return

            name = input("Passenger Name: ")
            age = int(input("Age: "))
            mobile = input("Mobile (10 digits): ")

            if len(mobile) != 10 or not mobile.isdigit():
                print(" Invalid mobile number.")
                return

            # Assign seat number
            seat_no = 41 - route_info["seats"]
            route_info["seats"] -= 1

            # Generate ticket
            ticket_id = self.ticket_id
            self.tickets[ticket_id] = {
                "name": name,
                "age": age,
                "mobile": mobile,
                "route": route_info["route"],
                "price": route_info["price"],
                "seat": seat_no
            }

            print(f"‚úÖ Ticket booked! Ticket ID: {ticket_id}, Seat No: {seat_no}")
            self.ticket_id += 1
        except ValueError:
            print(" Invalid input. Please try again.")

    # View ticket
    def view_ticket(self):
        try:
            tid = int(input("Enter Ticket ID: "))
            if tid in self.tickets:
                t = self.tickets[tid]
                print("\nüé´ Ticket Details:")
                print(f"Ticket ID: {tid}")
                print(f"Name: {t['name']}, Age: {t['age']}, Mobile: {t['mobile']}")
                print(f"Route: {t['route']}, Seat: {t['seat']}, Price: ‚Çπ{t['price']}")
            else:
                print(" Ticket not found.")
        except ValueError:
            print(" Invalid input.")

    # Cancel ticket
    def cancel_ticket(self):
        try:
            tid = int(input("Enter Ticket ID to cancel: "))
            if tid in self.tickets:
                route_name = self.tickets[tid]["route"]

                # Find which route and free one seat
                for r in self.routes.values():
                    if r["route"] == route_name:
                        r["seats"] += 1
                        break

                del self.tickets[tid]
                print(" Ticket cancelled successfully.")
            else:
                print(" Ticket not found.")
        except ValueError:
            print(" Invalid input.")


# ===== MAIN PROGRAM =====
system = BusReservation()

while True:
    print("\n--- Bus Reservation System ---")
    print("1. Show Available Routes")
    print("2. Book Ticket")
    print("3. View Ticket")
    print("4. Cancel Ticket")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        system.show_routes()
    elif choice == "2":
        system.book_ticket()
    elif choice == "3":
        system.view_ticket()
    elif choice == "4":
        system.cancel_ticket()
    elif choice == "5":
        print(" Thank you for using Bus Reservation System.")
        break
    else:
        print(" Invalid choice.")
```
